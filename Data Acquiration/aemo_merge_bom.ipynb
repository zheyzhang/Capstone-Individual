## Merge data set

```python
aemo = pd.read_csv("combined_csv-operational-demand.csv")
aemo[['Date','Trading Interval Time']] = aemo['Trading Interval'].str.split(" ",expand=True) 
# Split the column 'Trading Interval'
agg_aemo = pd.DataFrame(aemo.groupby(['Date'])['Operational Demand (MW)'].sum().reset_index())
# Sum the OD based on the Date in 'Trading Interval'
agg_aemo
## agg_aemo.to_csv('agg_aemo.csv',sep=',',index = False) # output the csvfile
```

```python
bom = pd.read_csv("perth_airport.csv")
bom
```

```python
merge_data = bom.merge(agg_aemo,on='Date',how = 'inner')
merge_data
```
